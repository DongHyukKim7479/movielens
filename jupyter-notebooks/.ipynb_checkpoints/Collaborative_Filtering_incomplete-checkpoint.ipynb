```python
def compute_adj_cossim(mat_spar, i):
    mat_row_means = mat_spar.mean(axis=1)
    
    np_idx_nonzero_user = mat_spar.nonzero()[0]
    np_idx_nonzero_item = mat_spar.nonzero()[1]
    
    np_sparse = np.dstack((np_idx_nonzero_user, np_idx_nonzero_item))[0, :, :]
    df_sparse_exist_rating = pd.DataFrame(np_sparse, columns=['user_id', 'item_id'])
    
    df_group_by_item = df_sparse_exist_rating.groupby('item_id')['user_id'].apply(np.array).reset_index(name='user_ids')
    np_iter = np.array(df_group_by_item['item_id'])
    
    np_ids_user_1 = df_group_by_item[df_group_by_item['item_id'] == i]['user_ids'].values[0]
    sum_prod_two_items = sum_squre_item1 = sum_squre_item2 = 0
    for j in np_iter:
        np_result = np.zeros(mat_spar.shape[1])

        np_ids_user_2 = df_group_by_item[df_group_by_item['item_id'] == j]['user_ids'].values[0]
        np_ids_user = np.unique(np.concatenate((np_ids_user_1, np_ids_user_2)))

        for user_id in np_ids_user:
            val_i_col = mat_spar[user_id, i]
            val_j_col = mat_spar[user_id, j]
            mean_i_row = mat_row_means[user_id].item()
            
            if val_i_col == 0:
                adj_val_i_col = 0
            else:
                adj_val_i_col = val_i_col - mean_i_row
                
            if val_j_col == 0:
                adj_val_j_col = 0
            else:
                adj_val_i_col = val_j_col - mean_i_row 

            prod_two_items = (adj_val_i_col) * (adj_val_i_col)
            squre_item1 = (adj_val_i_col) ** 2
            squre_item2 = (adj_val_i_col) ** 2  
            
            sum_prod_two_items += prod_two_items 
            sum_squre_item1 += squre_item1
            sum_squre_item2 += squre_item2

        den = (sum_squre_item1 ** 0.5) * (sum_squre_item2 ** 0.5)
        if den != 0:
            np_result[j] = (sum_prod_two_items / den)
        else:
            np_result[j] = 0

    return (i, np_result)

```

```python
from multiprocessing import Pool
from functools import partial
import time

mat_spar_train = sparse.csr_matrix(df_spar_train)

np_idx_nonzero_user = mat_spar_train.nonzero()[0]
np_idx_nonzero_item = mat_spar_train.nonzero()[1]

np_sparse = np.dstack((np_idx_nonzero_user, np_idx_nonzero_item))[0, :, :]
df_sparse_exist_rating = pd.DataFrame(np_sparse, columns=['user_id', 'item_id'])

df_group_by_item = df_sparse_exist_rating.groupby('item_id')['user_id'].apply(np.array).reset_index(name='user_ids')
np_iter = np.array(df_group_by_item['item_id'])

def work(df_spar, id_item):
    return id_item, compute_adj_cossim(mat_spar_train, id_item)

time_start = time.time()

p = Pool(processes=12)
iterable = np_iter
func = partial(work, df_spar_train)
result_pred = p.map(func, iterable)
p.close()

time_end = time.time()

time_process = time_end - time_start
print(time_process)
```

```python
def compute_two_items_sim(df_spar, i, j):
    series_row_means = df_spar.mean(axis=1)
    sum_prod_two_items = sum_squre_item1 = sum_squre_item2 = 0
    for k, row in df_spar.iloc[:, [i, j]].iterrows():
        val_i_col = row.iloc[0]
        val_j_col = row.iloc[1]
        mean_k_row = series_row_means[k]

        prod_two_items = (val_i_col - mean_k_row) * (val_j_col - mean_k_row)
        squre_item1 = (val_i_col - mean_k_row) ** 2
        squre_item2 = (val_j_col - mean_k_row) ** 2

        sum_prod_two_items += prod_two_items
        sum_squre_item1 += squre_item1
        sum_squre_item2 += squre_item2

    den = (sum_squre_item1 ** 0.5) * (sum_squre_item2 ** 0.5)
    if den != 0:
        print(sum_prod_two_items / den)
    else:
        print('den is zero. can\'t devide it at index {}, {}'.format(i, j))
    
    
    
    
    

```
